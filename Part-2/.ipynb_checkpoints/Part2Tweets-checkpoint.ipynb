```python
import matplotlib.pyplot as plt
import tweepy
import numpy as np
import pandas as pd
import datetime
import matplotlib.ticker as mtick
import seaborn as sns
now = datetime.datetime.now()

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import TW_consumer_key, TW_consumer_skey, TW_access_token, TW_access_stoken

# # Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(TW_consumer_key, TW_consumer_skey)
auth.set_access_token(TW_access_token, TW_access_stoken)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Target User Accounts
target_user = ("@BBC", "@CBS", "@CNN", "@Foxnews", "nytimes")

# List for dictionaries of results
results_list = []
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
screenname = []
tweettext= []
tweetdate = []
tweetnumber = []

# Loop through each user
for user in target_user:
    count=0
    # Loop through 5 pages of tweets (total 100 tweets)
    for x in range(1, 6):

        # Get all tweets from home feed
        public_tweets = api.user_timeline(user, page=x)
        
        # Loop through all tweets
        for tweet in public_tweets:
            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]

            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
            screenname.append(tweet["user"]["screen_name"])
            tweettext.append(tweet["text"])
            tweetdate.append(tweet["created_at"])
            count=count+1
            tweetnumber.append(count)
        
    # Create a dictionaty of results
user_results = {
    "Screen Name": screenname,
    "Date": tweetdate,
    "Tweet Text": tweettext,
    "Compound Score": compound_list,
    "Postive Score": positive_list,
    "Neutral Score": neutral_list,
    "Negative Score": negative_list,
    "Tweet Number": tweetnumber
}

```

```python
# Create DataFrame from Results List
results_df = pd.DataFrame.from_dict(user_results).round(3)
results_df
```

```python
results_df.to_csv('part2Tweets.csv', index=False)
```

```python
# Create a scatter plot showing sentiment analysis for each tweet

scatter = sns.pairplot(x_vars=["Tweet Number"], y_vars=["Compound Score"], data=results_df, hue='Screen Name', height=7)
scatter.fig.suptitle(f"Sentiment Analysis of Media Tweets ({now.strftime('%m/%d/%Y')})", y=1.01)
scatter.set(xlabel='Tweets Ago', ylabel='Tweet Polarity')
scatter._legend.set_title("Media Sources")
scatter.savefig("Images/Tweet_scatter.png")
```

```python
# Store all Relevant Percent Changes into a Tuple
sentiment = results_df.groupby("Screen Name")['Compound Score'].mean()
sentiment_tuple = tuple(sentiment)
num_values = np.arange(len(sentiment))
# Splice the data between passing and failing drugs
passing = [x for x in sentiment if x >=0]
failing = [x for x in sentiment if x <0]

bar_color = []

for x in sentiment:
    if x>0:
        bar_color.append('g')
    else:
        bar_color.append('r')

# Orient widths. Add labels, tick marks, etc. 

plt.figure(figsize=(8,8))
ax = sentiment.plot(kind='bar', x=num_values, y=sentiment_tuple, color=bar_color, align="center", width=.9)
ax.set_title(f"Overall Media Sentiment based on Twitter ({now.strftime('%m/%d/%Y')})")
ax.set_ylabel("Tweet Polarity")
ax.set_xlabel("Media Sources")
plt.grid(b=True, linewidth=.5, alpha=.5)


# Use functions to label the percentages of changes
def autolabel(ax):
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom
    for rect in ax.patches:
        height = rect.get_height()
    
        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

        # If we can fit the label above the column, do that;
        # otherwise, put it inside the column.
        if p_height > 0:
            label_position = y_bottom + abs(y_bottom) + .01
        else:
            label_position = y_bottom + abs(y_bottom) - .015

        ax.text(rect.get_x() + rect.get_width()/2., label_position,
                round(height,3), ha='center', va='bottom') 

# Call functions to implement the function calls
autolabel(ax)

# Save the Figure
plt.autoscale(enable=True, axis='x')
plt.savefig("Images/Tweet_bars.png", bbox_inches='tight')

# Show the Figure
plt.show()
```
